```python
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
iris = load_iris()
```

```python
# split the data into two parts, 1 for training and 1 for testing 
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=1337)
```

```python
# dimensions of X_train and X_test
X_train.shape, X_test.shape
```

```python
# pair plot to visualize the data
fig, ax = plt.subplots(3,3, figsize=(15,15))
plt.suptitle('iris_pairplot')
for i in range(3):
    for j in range(3):
        ax[i,j].scatter(X_train[:, j], X_train[:, i+1], c=y_train, s=60)
        ax[i, j].set_xticks(())
        ax[i, j].set_yticks(())
        if i == 2:
            ax[i, j].set_xlabel(iris['feature_names'][j])
        if j == 0:
            ax[i, j].set_ylabel(iris['feature_names'][i+1])
        if j>i:
            ax[i, j].set_visible(False)
```

```python
# import the knn classifier, and configure it for 1 neighbors
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# train or fit the model
knn.fit(X_train, y_train)
```

```python
# prredict for a value
X_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(X_new)
print(iris['target_names'][prediction])
```

```python
# compute the accuracy
y_prediction = knn.predict(X_test)
np.mean(y_prediction == y_test)
knn.score(X_test, y_test)
```
